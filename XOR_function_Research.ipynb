```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
NUMBER = 25

num = np.arange(NUMBER)
numArray1 = np.tile(num, (NUMBER,1))

numArray2 = np.reshape(num, (-1,1))
numArray2 = np.tile(numArray2, (1, NUMBER))
```

```python
xor_out = np.bitwise_xor(numArray1, numArray2)
```

```python
fig = plt.figure(figsize=(NUMBER*1, NUMBER*0.8))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

colors = plt.get_cmap('tab10').colors
for i in range(NUMBER):

    #change subplot to fit to your number
    ax = fig.add_subplot(5,5, i+1)
    ax.plot(num, xor_out[i].flatten('F'))
    ax.set_title('XOR of {}'.format(i))
plt.show()
```
